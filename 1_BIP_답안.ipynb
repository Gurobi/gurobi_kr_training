Binary Integer Program

\begin{align}
        \text{max }x+ y+ 2z \\
        x + 2y + 3z \le 4 \\
        x + y \ge 1 \\
        x,y,z \in \{0, 1\}
\end{align}

**gurobipy 설치 및 패키지 불러오기**

```python
!pip install gurobipy

import gurobipy as gp
from gurobipy import GRB
```

**모델 선언**

```python
model = gp.Model("myBIP")
```

**결정 변수**

```python
x = model.addVar(vtype=GRB.BINARY, name="x")
y = model.addVar(vtype=GRB.BINARY, name="y")
z = model.addVar(vtype=GRB.BINARY, name="z")
```

**목적 함수**

```python
model.setObjective(x + y + 2*z, GRB.MAXIMIZE)
```

**제약 조건**

```python
model.addConstr(x + 2*y + 3*z <= 4, name="Constraint1")
model.addConstr(x + y >= 1, name="Constraint2")
```

**Gurobi Optimizer 실행**

```python
model.optimize()
```

**최적화 상태 확인**

```python
print(f"The solution status is {model.status}.")
if model.status == GRB.OPTIMAL:
  print("The problem solved to optimality.")
```

**해 확인**

```python
for var in model.getVars():
    if var.x > 0.5:
        print(f"Variable name: {var.varName}; Optimal value: {var.x}")
```
